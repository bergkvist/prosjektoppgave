```python
pressure = pd.read_csv('./src/data/pipepressure.csv', index_col=0)
path = pd.read_csv('./src/data/well_path.csv', sep=';')
height = interpolate.interp1d(np.array(path.Md),np.array(path.Tvd))
```

```python
def diffs(arr):
    n = []
    for i in range(1, len(arr)):
        n.append(arr[i] - arr[i-1])
    return n

x = list(pressure.columns.map(float))
plt.plot(diffs(x))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pressure = pd.read_csv('./src/data/pipepressure.csv', index_col=0)
plt.rcParams["figure.figsize"] = (30,15)
sns.heatmap(pressure.transpose())
```

```python
p = pressure[20:205].transpose()

sns.heatmap(p)
```

```python
p = pressure[:].transpose()
params = {
    'cmap': 'inferno',
    'vmin': -4,
    'vmax': 4
}
rho_times_g = 0.1065 # Found empirically by trying values and observing the change in the heatmap
s = pd.Series(rho_times_g * height(np.array(p.index.map(float))), p.index)
n = p.subtract(s, axis='rows')
sns.heatmap(n, **params)
plt.imsave('n.png', n, **params)
```
